```python
import pandas as pd
import numpy as np
```

```python
movie_names=pd.read_csv('movies.csv')
movie_names.head()
```

```python
ratings=pd.read_csv('ratings.csv')
ratings.head()
```

```python
movie_data=pd.merge(ratings,movie_names,on='movieId')
movie_data.head()
```

```python
movie_data.info()
```

# **Criteria for Popularity Based Reccomandation System**

# 1.Movies with highest ratings
# 2.Number of views

```python
# Average Rating of movies
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
```

```python
rating_mean_count=pd.DataFrame(movie_data.groupby('title')['rating'].mean())
rating_mean_count['rating_count']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
```

```python
rating_mean_count.head()
```

```python
rating_mean_count['rating']=round(rating_mean_count['rating'],1)
# round off the rating column
```

```python
rating_mean_count.head()
```

```python
#select the movies that ratings are greater than 3
rating_mean_count = rating_mean_count[(rating_mean_count['rating']>3) & (rating_mean_count['rating_count']>100) ]
```

```python
rating_mean_count.head()
```

```python
rating_mean_count.shape
```

```python
#sort the ratings
rating_mean_count=rating_mean_count.sort_values('rating',ascending=False)
```
